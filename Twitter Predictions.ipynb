```python
def bar_chart(feature):
    tweetpre = train2[train2['retweeted']==1][feature].value_counts()
    tweetsen = train2[train2['retweeted']==0][feature].value_counts()
    df = pd.DataFrame([tweetpre, tweetsen])
    df.index = ['Goodtweets','NotSoGoodtweet']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('entities')
```
